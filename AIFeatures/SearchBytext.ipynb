```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import sigmoid_kernel
from fastapi import FastAPI, HTTPException
from pydantic import BaseModel
import logging

logging.basicConfig(level=logging.INFO)

# Step 1: Read the CSV file into a DataFrame
file_path = '/content/updated_products.csv'
try:
    main_data = pd.read_csv(file_path, on_bad_lines='skip')
except Exception as e:
    logging.error(f"Failed to read CSV file: {e}")
    raise

# Step 2: Take a subset of the data (first 15000 rows in this example)
subset_data = main_data.head(15000)

# Step 3: Drop rows with NaN values in 'articleType' and 'baseColour'
df_cleaned = subset_data.dropna(subset=['material', 'colors'])

# Combine 'articleType' and 'baseColour' into a new column 'combined'
df_cleaned.loc[:, 'combined'] = df_cleaned['colors'] + ' ' + df_cleaned['material']

# Initialize the TF-IDF Vectorizer
tfv = TfidfVectorizer()

# Fit and transform the 'combined' column
tfv_matrix = tfv.fit_transform(df_cleaned['combined'])  # For getting cosine similarity

# Compute sigmoid kernel for the matrix
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
sig_list = sig.tolist()  # Recommendation of similar products based on title

# Create an index mapping for the combined column
indices = pd.Series(df_cleaned.index, index=df_cleaned['combined']).drop_duplicates()

def recommendProduct(title, sig=sig_list):
    try:
        # Search for combined baseColour and articleType that match part of the title
        matching_titles = df_cleaned[df_cleaned['combined'].str.contains(title, case=False, na=False)]
        if matching_titles.empty:
            return []

        # Get the full names and their indices of the matching products
        matching_indices = matching_titles.index.tolist()
        product_ids = []
        for index in matching_indices:
            sig_cs = list(enumerate(sig[index]))
            sig_sorted = sorted(sig_cs, key=lambda x: x[1], reverse=True)
            sig_cs = sig_sorted[1:11]
            product_indices = [i[0] for i in sig_cs]
            product_ids.extend([str(df_cleaned.iloc[i]['productID']) for i in product_indices])  # Convert each ID to string

        # Remove duplicates and keep the first n_neighbors recommendations
        product_ids = list(dict.fromkeys(product_ids))[:10]
        return product_ids
    except Exception as e:
        logging.error(f"Error in recommendProduct: {e}")
        return []
```
